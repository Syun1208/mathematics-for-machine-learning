Data cleaning

```python
new = pd.DataFrame({'a': [np.nan, 2, 3], 'b': [1, np.nan, 6]}, index=range(3))
```

```python
new
```

```python
new.dropna()
```

```python
new.fillna(0)
```

```python
new.info()
```

```python
new['b'] = new['b'].fillna(0).astype('int')
new['b']
```

```python
new.info()
```

```python
new['b'].iloc[-1]
```

```python
new['b'].iloc[-1] = 7
new['b']
```

```python
new['a'].fillna(0, inplace=True)
```

```python
new
```

```python
new.loc[3, 'a'] = 2
new.loc[3, 'b'] = 0
```

```python
new
```

```python
new.duplicated().value_counts()
```

```python
new.duplicated().value_counts(normalize=True).plot.pie(autopct="%.1f")
```
